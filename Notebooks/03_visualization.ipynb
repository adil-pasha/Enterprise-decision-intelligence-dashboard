# Enterprise Decision Intelligence Dashboard
## 03 - Visualization

```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("../data/Sample - Superstore.csv")
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```
df.groupby("Region")["Sales"].sum().plot(kind="bar",title="Revenue by Region")
plt.show()
```

```
df.groupby("Category")["Sales"].sum().plot(kind="bar",title="Revenue by Category")
plt.show()
```

```
df.groupby("Region")["Profit"].sum().plot(kind="bar",title="Profit by Region")
plt.show()
```

```
df.groupby("Segment")["Sales"].sum().plot(kind="pie",autopct="%1.1f%%",ylabel="")
plt.show()
```

```
monthly=df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly.index=monthly.index.astype(str)
monthly.plot(figsize=(10,4))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10).sort_values().plot(kind="barh")
plt.show()
```

```
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10).sort_values().plot(kind="barh")
plt.show()
```
