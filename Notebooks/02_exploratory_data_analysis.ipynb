# Enterprise Decision Intelligence Dashboard
## 02 - Exploratory Data Analysis

```
import pandas as pd
df=pd.read_csv("../data/Sample - Superstore.csv")
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```
df["Sales"].sum()
df["Profit"].sum()
df["Order ID"].nunique()
df["Customer ID"].nunique()
```

```
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

```
df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
```

```
df.groupby("Category")[["Sales","Profit"]].sum()
```

```
df.groupby("Segment")["Sales"].sum()
```

```
df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```
df.groupby("Product Name")["Profit"].sum().sort_values().head(10)
```

```
monthly=df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly
```
